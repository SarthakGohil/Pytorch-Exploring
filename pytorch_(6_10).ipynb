```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.sample(5)
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
X=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder

LE=LabelEncoder()

y_train=LE.fit_transform(y_train)
y_test=LE.transform(y_test)
```

```python
X_train_tensor=torch.tensor(X_train,dtype=torch.float32)
X_test_tensor=torch.tensor(X_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.float32)
y_test_tensor=torch.tensor(y_test,dtype=torch.float32)
```

```python
from torch.utils.data import Dataset,DataLoader,TensorDataset
```

```python
class for_ds(Dataset):
  def __init__(self,features,labels):
    super().__init__()
    self.features=features
    self.labels=labels

  def __len__(self):
    return len(self.features)

  def __getitem__(self,idx):
    return self.features[idx],self.labels[idx]
```

```python
train_ds=for_ds(X_train_tensor,y_train_tensor)
test_ds=for_ds(X_test_tensor,y_test_tensor)
```

```python
train_loader=DataLoader(train_ds,batch_size=32,shuffle=True)
test_loader=DataLoader(test_ds,batch_size=32,shuffle=True)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class myNN2(nn.Module):
  def __init__(self,l1_features):
    super().__init__()
    self.linear=nn.Sequential(
        nn.Linear(l1_features,1)
    )
    self.criterian=nn.BCEWithLogitsLoss()
    self.optimizer=optim.SGD(self.parameters(),lr=0.1)

  def forward(self,x):
    return self.linear(x)

  def loss(self,y_pred,y):
    return self.criterian(y_pred,y)
```

```python
apun=myNN2(X_train_tensor.shape[1])
```

```python
lr=0.1
epochs=10

for epoch in range(epochs):
  for batch_features,batch_labels in train_loader:

    y_pred=apun(batch_features)
    loss=apun.loss(y_pred,batch_labels.view(-1,1))
    apun.optimizer.zero_grad()
    loss.backward()
    apun.optimizer.step()

  print(f"epoch: {epoch}   loss:{loss}")
```

```python
apun.eval()
accuracy=[]

with torch.no_grad():
  for batch_features,batch_labels in test_loader:
    y_pred=apun(batch_features)
    y_pred=(y_pred>0.8).float()
    accuracy.append((y_pred.view(-1)==batch_labels).float().mean().item())

overall_accuracy=sum(accuracy)/len(accuracy)
print(f"overall accuracy: {overall_accuracy}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.model_selection import train_test_split as tts
```

```python

torch.manual_seed(42)
```

```python
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"using device : {device}")
```

```python
df=pd.read_csv("/content/fashion-mnist_train.csv")
df.head()
```

```python
X=df.drop(columns='label').values
y=df['label'].values
X=X/255.0
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
X_train_tensor=torch.tensor(X_train,dtype=torch.float32)
X_test_tensor=torch.tensor(X_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.long)
y_test_tensor=torch.tensor(y_test,dtype=torch.long)
```

```python
class myMNIST(Dataset):
  def __init__(self,features,labels):
    super().__init__()
    self.features=features
    self.labels=labels

  def __len__(self):
    return len(self.features)

  def __getitem__(self,idx):
    return self.features[idx],self.labels[idx]
```

```python
train_ds=myMNIST(X_train_tensor,y_train_tensor)
test_ds=myMNIST(X_test_tensor,y_test_tensor)
```

```python
# train_loader = DataLoader(train_ds,batch_size=128,shuffle=True)
# test_loader = DataLoader(test_ds,batch_size=128,shuffle=True)
```

```python
class MyNN3(nn.Module):
  def __init__(self,input_dim, output_dim,num_hidden_layer,neuron_per_layer,weight_decay,optim_name,lr, dropout_rate):
    super().__init__()
    linear=[]

    for i in range(num_hidden_layer):
      if i==0:
        linear.append(nn.Linear(input_dim,neuron_per_layer))
      else: linear.append(nn.Linear(neuron_per_layer,neuron_per_layer))
      linear.append(nn.BatchNorm1d(neuron_per_layer))
      linear.append(nn.ReLU())
      linear.append(nn.Dropout(p=dropout_rate))
    linear.append(nn.Linear(neuron_per_layer,output_dim))

    self.model=nn.Sequential(*linear)
    self.criterion = nn.CrossEntropyLoss()

    if(optim_name=="Adam"):
      self.optimizer = optim.Adam(self.parameters(),lr=lr,weight_decay=weight_decay)
    elif (optim_name=="SGD"):
      self.optimizer = optim.SGD(self.parameters(),lr=lr,weight_decay=weight_decay)
    elif (optim_name=="RMSprop"):
      self.optimizer = optim.RMSprop(self.parameters(),lr=lr,weight_decay=weight_decay)

  def forward(self,x):
    return self.model(x)

  def loss(self,y_pred,y):
    return self.criterion(y_pred,y)
```

```python
def objective(trial):
  num_hiden_layer=trial.suggest_int("num_hiden_layer",1,8)
  num_neuron_per_layer=trial.suggest_int("num_neuron_per_layer",8,128,step=8)
  epochs=trial.suggest_int("epochs",10,50,step=5)
  lr=trial.suggest_float("lr",1e-5,1e-1,log=True)
  dropout_rate=trial.suggest_float("dropout_rate",0.1,0.5,step=0.1) # Corrected typo from 'dropoout_rate'
  batch_size=trial.suggest_categorical("batch_size",[32,64,128])
  optim_name=trial.suggest_categorical("optimizer",["SGD","Adam","RMSprop"])
  weight_decay=trial.suggest_float("weight_decay",1e-5,1e-1,log=True)


  train_loader = DataLoader(train_ds,batch_size=batch_size,shuffle=True)
  test_loader = DataLoader(test_ds,batch_size=batch_size,shuffle=True)

  input_dim=784
  output_dim=10

  model=MyNN3(input_dim,output_dim,num_hiden_layer,num_neuron_per_layer,weight_decay,optim_name,lr, dropout_rate)
  model=model.to(device)

  for i in range(epochs):
    model.train()
    for batch_feature,batch_label in train_loader:
      batch_feature = batch_feature.to(device)
      batch_label = batch_label.to(device)


      y_pred=model(batch_feature)
      loss=model.loss(y_pred,batch_label)

      model.optimizer.zero_grad()
      loss.backward()
      model.optimizer.step()

    model.eval()
    accuracy=[]
    #this is for fashion mnist dataset and output has 10 classes.

    with torch.no_grad():
      for batch_feature,batch_label in test_loader:

        batch_feature = batch_feature.to(device, non_blocking=True)
        batch_label = batch_label.to(device, non_blocking=True)
        y_pred=model.forward(batch_feature)
        y_pred=torch.argmax(y_pred,dim=1)
        accuracy.append((y_pred==batch_label).float().mean().item())

    overall_accuracy=sum(accuracy)/len(accuracy)
    return overall_accuracy
```

```python
class myNN3(nn.Module):
  def __init__(self,input_dim,n_hidden, h_dim,output_dim):
    super().__init__()
    stack=[]

    for i in range(n_hidden):
      if(i==0):
        stack.append(nn.Linear(input_dim,h_dim))
      else:
        stack.append(nn.Linear(h_dim,h_dim))

      stack.append(nn.BatchNorm1d(h_dim))
      stack.append(nn.ReLU())
      stack.append(nn.Dropout(p=study.best_params['dropout_rate']))

    stack.append(nn.Linear(h_dim,output_dim))
    self.linear=nn.Sequential(*stack)


  def forward(self, x):
    return self.linear(x)
```

```python
model3=myNN3(784,study.best_params['num_hiden_layer'],study.best_params['num_neuron_per_layer'],10)
model3=model3.to(device)
```

```python
train_ds=myMNIST(X_train_tensor,y_train_tensor)
test_ds=myMNIST(X_test_tensor,y_test_tensor)


train_loader = DataLoader(train_ds,batch_size=study.best_params['batch_size'],shuffle=True)
test_loader = DataLoader(test_ds,batch_size=study.best_params['batch_size'],shuffle=True)
```

```python
lr=study.best_params['lr']
epochs=study.best_params['epochs']

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model3.parameters(),lr=study.best_params['lr'],weight_decay=study.best_params['weight_decay'])

model3.train()
for i in range(epochs):
  for batch_feature,batch_label in train_loader:

    batch_feature = batch_feature.to(device)
    batch_label = batch_label.to(device)

    optimizer.zero_grad()

    y_pred=model3(batch_feature)
    loss=criterion(y_pred,batch_label)

    loss.backward()
    optimizer.step()

  if i%10==0 :  print(f"epoch: {i}  , loss: {loss}")
```

```python
model3.eval()
accuracy=[]
#this is for fashion mnist dataset and output has 10 classes.

with torch.no_grad():
  for batch_feature,batch_label in test_loader:

    batch_feature = batch_feature.to(device)
    batch_label = batch_label.to(device)

    y_pred=model3.forward(batch_feature)
    y_pred=torch.argmax(y_pred,dim=1)
    accuracy.append((y_pred==batch_label).float().mean().item())

overall_accuracy=sum(accuracy)/len(accuracy)
print(f"overall accuracy: {overall_accuracy}")
```
